````python
import numpy as np
'''
top_cats = ['Requests for Snow Plowing','Snow Removal', 'Requests for Street Cleaning', 'Request for Pothole Repair', 'Misc. Snow Complaint']
for name, group in boston['CASE_TITLE'].groupby(pd.TimeGrouper(freq = "1D")):
    values = group.value_counts()
    values = pd.DataFrame(values)
    values['case title'] = values.index
    values['date'] = name
    top_5_by_day = top_5_by_day.append(values)
#top_5_by_day.columns = ['number calls', 'case title', 'date']
top_5_by_day
top_5 = top_5_by_day[top_5_by_day['case title'].isin(top_cats)]
#top_5.reset_index().drop(['index'], axis=1)
print top_5
'''

#sorry this is so long! I talked to Evimaria about it and it seems like Will and I didn't understand groupby() as well as we
#should, I'm going to try to make it prettier, but I doubt I'll have time to figure it out before this is due
#however, it works!
def counts_for_case(df):
    index=0
    for day, day_group in df.groupby(df.index.map(lambda datetime: datetime.date())): #groups the data frame by the date 
        #print len(boston.groupby(boston.index.map(lambda datetime: datetime.date())))
        days_cats = day_group.groupby('CASE_TITLE')
        days_cats = pd.DataFrame(days_cats.size())
        days_cats = days_cats.reset_index()
        days_cats.columns = ['CASE_TITLE', 'count']    
        for idx, row in days_cats.iterrows():
            if row.CASE_TITLE =='Snow Removal':
                daily_counts.loc[index]['Snow Removal'] = row['count']
            elif row.CASE_TITLE == 'Request for Snow Plowing':
                daily_counts.loc[index]['Request for Snow Plowing'] = row['count']
            elif row.CASE_TITLE == 'Requests for Street Cleaning':
                daily_counts.loc[index]['Requests for Street Cleaning'] = row['count']
            elif row.CASE_TITLE == 'Request for Pothole Repair':
                daily_counts.loc[index]['Request for Pothole Repair'] = row['count']
            elif row.CASE_TITLE == 'Misc. Snow Complaint':
                daily_counts.loc[index]['Misc. Snow Complaint'] = row['count']
        daily_counts.loc[index]['Day'] = day
        index +=1
    daily_counts.fillna(0, inplace = True)
    daily_counts.index=daily_counts['Day']
    return daily_counts


daily_counts = pd.DataFrame(index=np.arange(0, len(boston.groupby(boston.index.map(lambda datetime: datetime.date())))), 
                          columns =['Request for Snow Plowing', 'Snow Removal', 
                                    'Requests for Street Cleaning', 'Request for Pothole Repair', 'Misc. Snow Complaint',
                                    'Day'])

counts_for_case(boston)
daily_counts = daily_counts.drop(['Day'], axis =1)
ig, axes = plt.subplots()
daily_counts.plot(ax=axes)


axes.set_title("Call Requests by Day")
axes.set_ylabel("# Calls")
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
snow_data.head(7)
#opened the csv in excel after getting some errors, there are -9999 values in both columns, need to get rid of those
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here
no_neg=snow_data[snow_data['SNOW']>=0] #make a variable that contains a data frame with values for SNOW of >= 0
no_neg.index = no_neg['DATE'] #set index to the 'date' column
#agg_mean = no_neg.groupby(pd.TimeGrouper(freq = "1D")).mean()
mean_snow=no_neg.groupby(no_neg.index.map(lambda datetime: datetime.date())).mean() #groupby object of the noerror data frame
#which uses the date to group and takes the mean of the snowfall for that date
mean_snow['SNOW']=mean_snow['SNOW'].apply(lambda x: x/10) #divides the mean in the snow column by 10 (for centimeters)
mean_snow.columns = ['SNWD', 'Avg. Snowfall']
axes =plt.subplot()
mean_snow.ix[:,1].plot(ax = axes, legend=True)

axes.set_ylabel('Snowfall in cm')
axes.set_title('Average Snowfall by Date')
axes.set_xlabel('Date')
````
