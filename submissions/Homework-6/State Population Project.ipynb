```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import vincent #rapid mapping tool
from vincent import AxisProperties, PropertySet, ValueRef #allows for a few different properties to be set with vincent
vincent.core.initialize_notebook() #necessary to activate functionality within ipython notebooks
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import sklearn.utils as utils
import sklearn.linear_model as linear_model
import sklearn.cross_validation as cross_validation
import sklearn.cluster as cluster
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from patsy import dmatrices

%matplotlib inline 
```

```python
data = pd.read_table("population_data_2014.csv", sep = ',', parse_dates = True, converters={'COUNTY': lambda x: str(x), 'STATE': lambda x: str(x)})
#the lambda function in the above is there because if you don't include it, it gets rid of the leading zeroes in the Fips Codes when it converts
#them to numpy numbers, I actually added the zeroes in excel because it was giving me issues to do so in python.
state_only = pd.DataFrame(columns = data.columns)
counties_only = pd.DataFrame(columns = data.columns)
for index, row in data.iterrows():
    if row.STNAME == row.CTYNAME and row.STNAME != "District of Columbia":
        #print "True!"
        state_only = state_only.append(row)
    elif row.STNAME != row.CTYNAME and row.STNAME != "District of Columbia":
        counties_only = counties_only.append(row)
```

```python
costs_state = pd.read_table('state_statistics_for_download.csv', sep = ',', parse_dates = True)
abbreviations = pd.read_table('Abbreviations', sep = ',')
state_sizes = pd.read_table('State_sizes.csv', sep = ',')

costs_state_fullname = pd.merge(costs_state, abbreviations, left_on = "State", right_on = 'Abb')
costs_state_fullname = pd.merge(costs_state_fullname, state_sizes, left_on = 'State_y', right_on = 'State Name')

templist = []
for index, row in costs_state_fullname.iterrows():
    if row['Year-Quarter'] == '2010Q1':
        templist.append(row.tolist())

costs_state_fullname = pd.DataFrame(templist)
costs_state_fullname = costs_state_fullname[[1,2,3,4,5,8]]
costs_state_fullname.columns = ['YQ', 'AVEPRICE', 'MEDPRICE', 'STATE', 'ABB', 'SQMILES']
stonly_all = pd.merge(state_only, costs_state_fullname, left_on = 'STNAME', right_on = 'STATE')
stonly_all
#this data frame now has data for everything the original data set had, as well as information about the median price of a home
#in that state, and the size of that state in square miles
```

```python
counties_only['FIPS'] = counties_only.STATE + counties_only.COUNTY #this creates a complete FIPS code from the previous
```

```python
state_borders = r'us_states.topo.json'
states_data = [{'name': 'states',
             'url': state_borders,
             'feature': 'us_states.geo'}]
states_pop = vincent.Map(data=stonly_all, geo_data=states_data, scale=1500, projection='albersUsa',
          data_bind='CENSUS2010POP', data_key='STNAME', 
          map_key={'states': 'properties.NAME'})
states_pop.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.85)
states_pop.rebind(column = 'CENSUS2010POP', brew = 'OrRd')
states_pop.to_json('States_Population_choropleth.json', html_out=True, html_path='states_Population_choropleth.html')
states_pop.scales['color'].type = 'threshold'
states_pop.scales['color'].domain = [0, 2000000, 4000000, 6000000, 8000000, 10000000, 12000000]
states_pop.legend('Population')
states_pop.display()
#When looking at this map it seems that California, Texas, Florida, Illinois, New York, Ohio, and Pennsylvania all have the
#highest populations, and might be good places to invest in housing
```

```python
state_borders = r'us_states.topo.json'
states_data = [{'name': 'states',
             'url': state_borders,
             'feature': 'us_states.geo'}]
states_cost = vincent.Map(data=stonly_all, geo_data=states_data, scale=1500, projection='albersUsa',
          data_bind='MEDPRICE', data_key='STNAME', 
          map_key={'states': 'properties.NAME'})
states_cost.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.85)
states_cost.rebind(column = 'MEDPRICE', brew = 'OrRd')
states_cost.to_json('States_Population_choropleth.json', html_out=True, html_path='states_Population_choropleth.html')
states_cost.scales['color'].type = 'threshold'
states_cost.scales['color'].domain = [0, 50000, 100000, 150000, 200000, 250000, 300000]
states_cost.legend('Median Price')
states_cost.display()
```

```python
county_borders = r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_borders,
             'feature': 'us_counties.geo'}]
choro_pop = vincent.Map(data=counties_only, geo_data=geo_data, scale=1500, projection='albersUsa',
          data_bind='RINTERNATIONALMIG2014', data_key='FIPS',
          map_key={'counties': 'properties.FIPS'})
choro_pop.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.5)
choro_pop.rebind(column = 'ESTIMATESBASE2010', brew = 'OrRd')
choro_pop.to_json('Counties_Population_choropleth.json', html_out=True, html_path='Counties_Population_choropleth.html')
choro_pop.scales['color'].type = 'threshold'
choro_pop.scales['color'].domain = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000]
choro_pop.legend('Population')
choro_pop.display()

#when you look at counties, however, it's clear that the population density picture is very different than the larger map would
#indicate. In fact, populations are densely distributed in very small areas, which seem to be expanding (see the darker
#counties near the densely populated counties)
```
