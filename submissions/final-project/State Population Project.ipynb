```python
data = pd.read_table("population_data_2014.csv", sep = ',', parse_dates = True, converters={'COUNTY': lambda x: str(x), 'STATE': lambda x: str(x)})
#the lambda function in the above is there because if you don't include it, it gets rid of the leading zeroes in the Fips Codes when it converts
#them to numpy numbers, I actually added the zeroes in excel because it was giving me issues to do so in python.
state_only = pd.DataFrame(columns = data.columns)
counties_only = pd.DataFrame(columns = data.columns)
for index, row in data.iterrows():
    if row.STNAME == row.CTYNAME and row.STNAME != "District of Columbia":
        #print "True!"
        state_only = state_only.append(row)
    elif row.STNAME != row.CTYNAME and row.STNAME != "District of Columbia":
        counties_only = counties_only.append(row)
```

```python
def trendline(xd, yd, order=1, c='r', alpha=1, Rval=False):
    """Make a line of best fit"""

    #Calculate trendline
    coeffs = np.polyfit(xd, yd, order)

    intercept = coeffs[-1]
    slope = coeffs[-2]
    if order == 2: power = coeffs[0]
    else: power = 0

    minxd = np.min(xd)
    maxxd = np.max(xd)

    xl = np.array([minxd, maxxd])
    yl = power * xl ** 2 + slope * xl + intercept

    #Plot trendline
    plt.plot(xl, yl, c, alpha=alpha)

    #Calculate R Squared
    p = np.poly1d(coeffs)

    ybar = np.sum(yd) / len(yd)
    ssreg = np.sum((p(xd) - ybar) ** 2)
    sstot = np.sum((yd - ybar) ** 2)
    Rsqr = ssreg / sstot

    if Rval == False:
        #Plot R^2 value
        plt.text(0.8 * maxxd + 0.2 * minxd, 0.8 * np.max(yd) + 0.2 * np.min(yd),
                 '$R^2 = %0.2f$' % Rsqr)
    else:
        #Return the R^2 value:
        return Rsqr
```

```python
def plot_with_r2(x, y, Name_x, Name_y, subplot_location):
    title = Name_x + ' vs. ' + Name_y
    plt.subplot(subplot_location)
    plt.scatter(x, y)
    plt.title(title, fontsize = 12)
    plt.ylabel(Name_y, fontsize = 12)
    plt.xlabel(Name_x, fontsize = 12)
    rsqr = trendline(x,y, order = 1, c ='r', alpha = 1, Rval = True)
    rsqr = np.around(rsqr, decimals = 4)
    rsqr = 'R^2 = ' + str(rsqr)
    plt.text(0, 0, rsqr, fontsize = 12, bbox=dict(alpha=0.25))
```

```python
fig, ax = plt.subplots(3,3, figsize = (15, 15))
plot_with_r2(analysis_set.CENSUS2010POP, analysis_set.MEDPRICE, "Population Size", 'Median Price', 331)
plot_with_r2(analysis_set.NPOPCHG_2010, analysis_set.MEDPRICE, 'Population Change', 'Median Price', 332)
plot_with_r2(analysis_set.BIRTHS2010, analysis_set.MEDPRICE, 'Births', 'Median Price', 333)
plot_with_r2(analysis_set.DEATHS2010, analysis_set.MEDPRICE, 'Deaths', 'Median Price', 334)
plot_with_r2(analysis_set.NETMIG2010, analysis_set.MEDPRICE, 'Migration', 'Median Price', 335)
plot_with_r2(analysis_set.SQMILES, analysis_set.MEDPRICE, 'State Size', 'Median Price', 336)
plot_with_r2(analysis_set.AVEPRICE, analysis_set.MEDPRICE, 'Average Price', 'Median Price', 337)
plot_with_r2(analysis_set.MEDPRICE, analysis_set.MEDPRICE, 'Median Price', 'Median Price', 338)
plot_with_r2(analysis_set.POPDENSITY, analysis_set.MEDPRICE, 'Population Density', 'Median Price', 339)
plt.tight_layout()
```
