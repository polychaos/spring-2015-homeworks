```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import time
import seaborn as sns
import math
import scipy.cluster.hierarchy as hr
import matplotlib.pyplot as plt
import time
import scipy as sp
import sklearn.metrics as metrics
import scipy.sparse.linalg as linalg
import sklearn.cluster as cluster
from sklearn.decomposition import TruncatedSVD
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
menu = pd.io.parsers.read_table("menu_recipes.txt", sep = '\t', header = None)
epic = pd.io.parsers.read_table("epic_recipes.txt", sep = '\t', header = None)
allr = pd.io.parsers.read_table("allr_recipes.txt", sep = '\t', header = None)
map_names = pd.io.parsers.read_table("map.txt", sep = '\t', header = None)
all_sources = [menu, epic, allr]
all_recipes = pd.concat(all_sources, keys = ['menu', 'epic', 'allr'])
#this just opens the files as file objects, allowing us to iterate over them. can't print anything directly though
#I want to make a data frame that has all of the possible ingredients on the columns, and recipes on the rows. Simple binary
#data will represent the presence or absence of each of the ingredients in that recipe
all_columns = np.unique(all_recipes.ix[:,1:].values.ravel()) #I'm going to use these as my columns
menu_list = [i for i in all_columns]
unique_ingredients = np.unique(menu_list)
#menu_list = [i for i in menu[1]]
```

```python
print all_recipes.ix[0:2,0:8]
all_recipes_binary = pd.DataFrame(np.zeros((57691,384), dtype = np.int), columns = unique_ingredients)
```

```python
#this cell is for making a new data frame that isn't heirarchically labeled
all_recipes_oneindex = pd.concat(all_sources)
menu = pd.read_csv("menu_recipes.txt", sep = '\n', header = None)
epic = pd.read_csv("epic_recipes.txt", sep = '\n', header = None)
allr = pd.read_csv("allr_recipes.txt", sep = '\n', header = None)

from_all = pd.concat([menu, epic, allr])
from_all.info()
#I'm pulling in all of the different website's info, there may be some duplicates (e.g. chinese vs. China)
```

```python
from_all['Country'] = from_all[0].str.split('\t').str[0]
from_all['Ingred'] = from_all[0].str.split('\t').str[1:]

from_all = from_all.drop(0, axis =1)
from_all= from_all.set_index(['Country'])
from_all['Ingred']=from_all['Ingred'].apply(lambda row: " ".join(row))

region_column = []
all_countries = [i for i in from_all.index]
map_names = map_names.set_index(map_names.ix[:,0])
for i in all_countries:
    region_column.append(map_names.ix[i, 1])
from_all['region'] = region_column
from_all = from_all.set_index(from_all.region)
from_all = from_all.drop("region", axis = 1)
from_all

#changing the formatted text in the ingredients column to a block of text, also setting the index to be the country
#and dropping the earlier column
#it also changes the country column to a region column to reduce the number of clusters
```

```python
country_index=from_all.index.unique().tolist()

grouped_by_country=pd.DataFrame(index=country_index, columns=['Ingred'])
grouped_by_country =grouped_by_country.reset_index()

print np.unique(from_all.index) #grouped_by_country.index
for idx, row in grouped_by_country.iterrows():
    #ingredients = []
    #ingredients = from_all.ix[grouped_by_country['index'].ix[idx]]['Ingred'].tolist()
    row.Ingred = from_all.ix[grouped_by_country['index'].ix[idx]]['Ingred'].tolist()
    #row.Ingred = ingredients
grouped_by_country['Ingred']=grouped_by_country['Ingred'].apply(lambda row: " ".join(row))
grouped_by_country = grouped_by_country[grouped_by_country['index'] != 'American']
grouped_by_country.head()
regions = grouped_by_country['index'].tolist()
grouped_by_country.head()

#effectively concatenates each of the rows into one row for each index, making the block of text larger for each row
#I was originally thinking groupby for this, but it's hard (impossible?) to pass a groupby object into a vectorizer
```

```python
vectorizer = TfidfVectorizer(min_df=1, max_df=.99)
for_all_vec = vectorizer.fit_transform(grouped_by_country['Ingred'])
```

```python
top_ingred= vectorizer.get_feature_names()
distanceMatrix = pdist(for_all_vec.todense(),'euclidean')
Z = hr.linkage(distanceMatrix,  method='complete', metric='euclidean')
print Z.shape
```

```python
fig = plt.figure(figsize=(20,20))
T = hr.dendrogram(Z,color_threshold=1.0, labels =regions, leaf_font_size=10)
fig.show()
```

as can be seen in the above dendrogram, there are clear connections between the different types of cuisine which relate relatively well to their geographic location (or history).

Everyone is clustered within the european lineage, which is either a testament to imperial colonialism OR the desire for novelty in the european diet. On the opposite end of the scale are South Asians and Africans, who presumably share a number of spices.
